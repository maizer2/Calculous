```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x**2
```

```python
def df(x):
  return 2*x
```

```python
# np.linespace는 지정된 범위 내에서 균등한 간격으로 원하는 개수만큼의 숫자를 생성
# x는 -10 ~ 10까지 400개의 실수들
x = np.linspace(-10, 10, 400)
y = f(x)
dydx = df(x)
```

```python
x0 = 2
y0 = f(x0)
slope = df(x0)

tangent = slope * (x - x0) + y0
```

```python
slope
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2', color='blue')
plt.plot(x, dydx, label="f'(x) = 2x (도함수)", color='green', linestyle='--')
plt.plot(x, tangent, label=f"Tangent at x={x0}", color='red', linestyle=':')
```

```python
# f(x) 시각화
plt.plot(x, y, label="f(x) = x^2")
# plt.scatter(x, y, label="f(x) = x^2")
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='gray', linestyle="--")
plt.axvline(0, color="gray", linestyle="--")
plt.grid(True)
plt.legend()
plt.show()
```
