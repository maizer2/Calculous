```python
import numpy as np
import matplotlib.pyplot as plt
```

$$f(x)=x^2$$

```python
def f(x):
  return x**2
```

$$df(x)=f'(x)=2^x$$

```python
def df(x):
  return 2*x
```

```python
# np.linespace는 지정된 범위 내에서 균등한 간격으로 원하는 개수만큼의 숫자를 생성
# x는 -10 ~ 10까지 400개의 실수들
x = np.linspace(-10, 10, 400)
y = f(x)
dydx = df(x)
```

직선 방정식
$$y = mx+b$$

$$ y-y_0 = slope * (x - x_0) $$
$$ y = slope * (x-x_0) + y_0 $$


```python
x0 = 3
y0 = f(x0)
slope = df(x0)

tangent = slope * (x - x0) + y0
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2', color='blue')
plt.plot(x, dydx, label="f'(x) = 2x (Derivatives)", color='green', linestyle='--')
plt.plot(x, tangent, label=f"Tangent at x={x0}", color='red', linestyle=':')

plt.scatter([x0], [y0], color='black', zorder=5)  # 점 (x0, y0) 표시
plt.text(x0, y0, f'({x0}, {y0})', ha="center")

plt.title("Differential visualization: functions, derivatives, tangents")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
